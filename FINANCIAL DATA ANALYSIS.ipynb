# WHOLE DATA FOR ANALSIS

```python
#Merging the selected data from the balance sheet and the income statement
wholedata = pd.concat([merged_balancesheet_data,merged_incomestatement_data])
wholedata
```

```python
# changing a column name from the data
change_column = np.array(wholedata.columns)
change_column[0] = 'S/N'
change_column
```

```python
# now lets try and receate the datafame with the new column
data = pd.DataFrame(wholedata.values, columns = change_column)
data.head()
```

```python
#Transposing the data columns and rows
data_transposed = data.T 
data_transposed
```

```python
data_transposed.columns = data_transposed.iloc[0]
data_transposed
```

```python
cleaned_fiancial_data = data_transposed.drop(labels='S/N',axis=0)
cleaned_fiancial_data
```

```python
cleaned_fiancial_data["Gross Margin"].plot()
plt.title("GROSS PROFIT OF NIO", fontsize =16, fontweight = "bold")
plt.ylabel("Gross Margin", fontsize = 13)
plt.xlabel("Year", fontsize = 13)
plt.show()


cleaned_fiancial_data["Operating Margin"].plot()
plt.title("OPERATING PROFIT OF NIO", fontsize =16, fontweight = "bold")
plt.ylabel("Operating Margin", fontsize = 13)
plt.xlabel("Year", fontsize = 13)
plt.show()

cleaned_fiancial_data["Profit Margin"].plot()
plt.title("PROFIT MARGIN OF NIO", fontsize =16, fontweight = "bold")
plt.ylabel("Profit Margin", fontsize = 13)
plt.xlabel("Year", fontsize = 13)
plt.show()
```

# LIQUIDITY RATIOS

#Liquidity ratio is used to to determine a companys ability to pay its short-term debt obligations.
#Here we would look at the current ratio and quick ratio

```python
current_asset = cleaned_fiancial_data['Total Current Assets']
current_asset
```

```python
current_liabilities = cleaned_fiancial_data['Total Current Liabilities']
current_liabilities
```

```python
inventory = cleaned_fiancial_data['Inventory']
inventory
```

```python
current_ratio = current_asset/current_liabilities
current_ratio
```

```python
quick_ratio = (current_asset - inventory)/current_liabilities
quick_ratio
```

# VISUALIZATION OF THE LIQUIDITY RATIOS

```python
plt.plot(current_ratio, 'y', label='Line y')
plt.title("CURRENT RATIO OF NIO", fontsize =16, fontweight = "bold")
plt.ylabel("Current ratio", fontsize = 13)
plt.xlabel("Year", fontsize = 13)
plt.show()
plt.savefig('CURRENT RATIO OF NIO.png')

plt.plot(quick_ratio, 'r', label='Line y')
plt.title("Quick RATIO OF NIO", fontsize =16, fontweight = "bold")
plt.ylabel("Quick ratio", fontsize = 13)
plt.xlabel("Year", fontsize = 13)
plt.show()
plt.savefig('Quick RATIO OF NIO.png')
```

# PART B : STOCK PRICES VERSUS INFLATION

```python
import pandas_datareader.data as web
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def cleandata(ticker):
    import pandas_datareader.data as web
    start='2020-01-01'
    end='2021-12-31'
    df = web.DataReader(ticker,'yahoo',start,end)
    df.dropna()
    df.rename(columns = {'Close' : ticker+'_Close'}, inplace = True)
    return df
```

```python
df_Tesla = cleandata('TSLA')
df_HMC = cleandata('HMC')
df_TM = cleandata('TM')
df_NIO = cleandata('NIO')
```

```python
df_Tesla.head()
```

```python
df_HMC.head()
```

```python
df1= df_Tesla['TSLA_Close']
df2= df_HMC['HMC_Close']
df3= df_TM['TM_Close']
df4= df_NIO['NIO_Close']
```

```python
combined = pd.concat([df1, df2,df3,df4], axis=1)
combined
```

```python
combined.plot(title='Stock prices', xlabel= "Date", ylabel="Stock prices($)")
```
